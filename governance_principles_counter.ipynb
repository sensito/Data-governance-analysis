```python
import re
from collections import defaultdict
from os import listdir
from os.path import isfile, join
import PyPDF2

# Definición de principios con expresiones regulares para ambos idiomas
principios = {
    'procesos': re.compile(r'\bprocesos\b|\bprocesses\b', re.IGNORECASE),
    'transparencia': re.compile(r'\btransparencia\b|\btransparency\b', re.IGNORECASE),
    'roles': re.compile(r'\broles\b', re.IGNORECASE),
    'seguridad': re.compile(r'\bseguridad\b|\bsecurity\b', re.IGNORECASE),
    'privacidad': re.compile(r'\bprivacidad\b|\bprivacy\b', re.IGNORECASE),
    'accesibilidad': re.compile(r'\baccesibilidad\b|\baccessibility\b', re.IGNORECASE),
    'calidad': re.compile(r'\bcalidad\b|\bquality\b', re.IGNORECASE),
    'políticas': re.compile(r'\bpolíticas\b|\bpolicies\b', re.IGNORECASE),
    'madurez': re.compile(r'\bmadurez\b|\bmaturity\b', re.IGNORECASE),
    'ciclo de vida de los datos': re.compile(r'\bciclo de vida de los datos\b|\bdata lifecycle\b|\bdata life cycle\b', re.IGNORECASE),
}

# Ruta al directorio donde se encuentran los PDFs
directorio_pdfs = './data/'
```

```python
from PyPDF2 import PdfFileReader
import os

def extraer_texto_pdf(archivo):
    texto = ""
    with open(archivo, 'rb') as f:
        lector_pdf = PdfFileReader(f)
        for pagina in range(lector_pdf.getNumPages()):
            texto += lector_pdf.getPage(pagina).extractText()
    return texto
```

```python
def extraer_texto(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            if page.extract_text():
                text += page.extract_text()
        return text

def contar_principios(directorio_pdfs, principios):
    conteo_global = defaultdict(int)
    conteos_por_documento = {}

    # Recorrer cada archivo en el directorio
    for filename in listdir(directorio_pdfs):
        if filename.endswith('.pdf'):
            pdf_path = join(directorio_pdfs, filename)
            texto = extraer_texto(pdf_path)
            conteos_documento = defaultdict(int)

            # Contar cada principio en el documento
            for nombre, regex in principios.items():
                conteo = len(regex.findall(texto))
                conteos_documento[nombre] += conteo
                conteo_global[nombre] += conteo

            # Guardar conteos por documento
            conteos_por_documento[filename] = dict(conteos_documento)

    return dict(conteo_global), conteos_por_documento

# Ejecutar la función y imprimir los resultados
conteo_global, conteos_por_documento = contar_principios(directorio_pdfs, principios)
print("Conteo Global:", conteo_global)
print("Conteos por Documento:", conteos_por_documento)
```
